```python
import pandas as pd
import numpy as np

depthprofile1 = "/homes/adrian/mdr_mtb-master/results/summary/samtools_depth/WGS_131P_S4/DepthProfile_WGS_131P_S4.csv"
abresprofile1 = "/homes/adrian/mdr_mtb-master/results/summary/ABres/WGS_131P_S4/tmp/ABres_WGS_131P_S4.csv"

df = pd.read_csv(depthprofile1, header = 0, sep = ',')
df2 = pd.read_csv(abresprofile1, header = 0, sep = '\t')

df2 = df2.drop(df2[df2.Read_depth < 8].index)

def convert_lists(value):
    try:
        value = int(value)
    except:
        value = sum(ast.literal_eval(value))
    return value

ls = []
gene_loci = df["Gene"].drop_duplicates().to_list()
for gene in gene_loci:
    gene_df = df.loc[df["Gene"] == gene]
    ls.append([gene,round(gene_df["Depth"].mean())])
df = pd.DataFrame(ls)
df=df.rename(columns={0:"Mut_gene",1:"AvLocus_Depth"})
res = df2.merge(df,on='Mut_gene')

locdepth = res.pop("AvLocus_Depth")
res.insert(11, "AvLocus_Depth", locdepth)
res['Alt_num'] = res['Alt_num'].apply(convert_lists)
res["Var_Qual"]=round(np.square(res["Alt_num"]/res["Read_depth"])*np.log10(res["Alt_num"]),2)
res
#res["Var_Qual"]=round(np.square(res["Alt_num"]/res["Read_depth"])*np.log10(res["Alt_num"]),2)
```
