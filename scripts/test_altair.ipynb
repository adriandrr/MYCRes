```python
import altair as alt
import pandas as pd
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
df = pd.read_csv("/homes/adrian/mdr_mtb-master/backup/results(-0)/qc/samtools_depth/1_S1/1_S1_coverage_summary.txt" , header= 0, sep='\t')
df2 = pd.read_csv("/homes/adrian/mdr_mtb-master/backup/results(-40)/qc/samtools_depth/1_S1/1_S1_coverage_summary.txt" , header= 0, sep='\t')
df = df.rename(columns={"#rname":"gene_loci"})
df2 = df2.rename(columns={"#rname":"gene_loci"})
```

```python
barplot1 = alt.Chart(df).mark_bar().encode(
    x="gene_loci",
   y=alt.Y("meandepth",scale=alt.Scale(domain=[0,25000])),
    color="coverage",
    tooltip=["meandepth","coverage"]
).properties(
    title="1_S1_coverage_-0"
)

barplot2 = alt.Chart(df2).mark_bar().encode(
    x="gene_loci",
    y=alt.Y("meandepth",scale=alt.Scale(domain=[0,25000])),
    color="coverage",
    tooltip=["meandepth","coverage"]
    
).properties(
    title="1_S1_coverage_-40"
)

barplot1 | barplot2
```

```python
import altair as alt
import pandas as pd
import warnings

interval = alt.selection_interval(encodings=['x'])

warnings.simplefilter(action='ignore', category=FutureWarning)
path = "/homes/adrian/mdr_mtb-master/backup/results(-0)/qc/samtools_depth/10_S4/loci_depth/depth_"
gyrA = pd.read_csv(path + "gyrA.txt", header = 0, sep = '\t').drop(["#CHROM"], axis = 1)
gyrA = gyrA.rename(columns={"POS":"gyrA_POS","results/mapped/10_S4.sorted.bam":"gyrA_depth"})
depth = alt.Chart(gyrA).mark_bar().encode(
    x="gyrA_POS:Q",
    y="gyrA_depth:Q",
    tooltip=["gyrA_POS:Q","gyrA_depth:Q"]
).properties(width=800,height=400,title="1_S1_depth_-0").add_selection(
    interval
)

path2 = "/homes/adrian/mdr_mtb-master/backup/results(-40)/qc/samtools_depth/10_S4/loci_depth/depth_"
gyrA2 = pd.read_csv(path2 + "gyrA.txt", header = 0, sep = '\t').drop(["#CHROM"], axis = 1)
gyrA2 = gyrA2.rename(columns={"POS":"gyrA_POS","results/mapped/10_S4.sorted.bam":"gyrA_depth"})
depth2 = alt.Chart(gyrA2).mark_bar().encode(
    x=alt.X("gyrA_POS:Q"),
    y="gyrA_depth:Q",
    tooltip=["gyrA_POS:Q","gyrA_depth:Q"]
).properties(width=800,height=400,title="1_S1_depth_-40").add_selection(
    interval
)

depth | depth2
```
